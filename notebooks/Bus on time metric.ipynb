Our project is to find the the best neighborhood in Pittsburgh, and the metric I have chosen to analyze is how punctual the Port Authority buses are. There were many factors that went into choosing this metric, but the main reason it was selected was because I thought that in an ideal world, the bus always shows up on time. I also thought that in area's where public transit is more reliable, it is more likely for there to be schools and businesses which rely on public transport. Generally, I find that area's with school's and businesses tend to be safer, have more recreational areas, and restaurants, all of which I think should be considered when looking for the "best neighborhood." Another reason I have to mention, is that bus routes by nature are tied to neighborhoods, which makes working with the data very easy. I primarily focused on community oriented metrics like how often trains are on time as well as the reliability of phones, but I ultimately decided that for Pittsburgh specifically, buses were the most relevant and accurate metric.

The Metric: https://data.wprdc.org/dataset/port-authority-monthly-average-on-time-performance-by-route/resource/00eb9600-69b5-4f11-b20a-8c8ddd8cfe7a
    
My metric, as stated above is the promptness of bus routes. In order to do this, the columns on my csv I will be measuring are the bus id, the route name, and the on time percentage. On top of that, I'll be looking at the mode of transportation to make sure that I don't accidentally include the T or light rails, which are also in this dataset for some reason.

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("0010FinalData.csv")

grouped = df.groupby('route_full_name')['on_time_percent'].mean().reset_index()

colors = []
for percentage in grouped['on_time_percent']:
    if percentage >= 0.8:
        colors.append('green')
    elif percentage >= 0.65:
        colors.append('yellow')
    else:
        colors.append('red')

plt.figure(figsize=(12, 30))
plt.barh(grouped['route_full_name'], grouped['on_time_percent'], color=colors, height = .8)

plt.title('Bus On-Time Performance by Neighborhood')
plt.xlabel('On-Time Percentage')
plt.ylabel('Neighborhood')

plt.show()
```
