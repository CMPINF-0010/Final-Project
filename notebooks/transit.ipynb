# Introduction:
Our project is to find the the best neighborhood in Pittsburgh, and the metric I have chosen to analyze is how punctual the public transport is. There were many factors that went into choosing this metric, but the main reason it was selected was because I thought that in an ideal world, the public transportation always shows up on time. I also thought that in area's where public transit is more reliable, it is more likely for there to be schools and businesses which rely on public transport. Generally, I find that area's with school's and businesses tend to be safer, have more recreational areas, and restaurants, all of which I think should be considered when looking for the neighborhood with the most money-making potential. Another reason I have to mention, is that public transportatoin routes by nature are tied to neighborhoods, which makes working with the data very easy. I initially focused on community oriented metrics like how density of restaurants, but ultimately these community orriented metrics were not the most useful so I settled on public transport.

The Metric: https://data.wprdc.org/dataset/port-authority-monthly-average-on-time-performance-by-route/resource/00eb9600-69b5-4f11-b20a-8c8ddd8cfe7a
    
My metric, as stated above is the promptness of public transportation. In order to do this, the columns on my csv I will be measuring are the the route name and the on time percentage.

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("../datasets/transit.csv")

grouped = df.groupby('route_full_name')['on_time_percent'].mean().nlargest(15).reset_index()

colors = []
for percentage in grouped['on_time_percent']:
    if percentage >= 0.82:
        colors.append('green')
    elif percentage >= 0.8:
        colors.append('yellow')
    else:
        colors.append('red')

plt.figure()
plt.barh(grouped['route_full_name'], grouped['on_time_percent'], color=colors, height = .5)

plt.title('Public Transport Performance by Neighborhood')
plt.xlabel('On-Time Percentage')
plt.ylabel('Neighborhood')

plt.show()
```

# The Best Neighborhood
The best neighborhood based on public transport reliability is Overbrook, which has the two most reliable modes of public transport both being on time over 80% of the time. Something to note is that had trains not been factored in, the best neighborhood would be Manchester, which also is on time 80% of the time.
